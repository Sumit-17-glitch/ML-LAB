```python
print(df.isna().sum());
df["Age"].fillna(df["Age"].mode()[0], inplace= True);
print(df.isna().sum());
```

```python

print(df[(df["Sex"] == "Male") & (df["Age"] >= 20)]);
```

```python
plt.scatter(df["Age"], df["Survived"], color= "red");
plt.xlabel("Age");
plt.ylabel("Survived");
plt.title(" Age VS Survived")
plt.show();
```

```python
x = df["Sex"].value_counts().index;
print(x);
y = df["Sex"].value_counts().values;
print(y);
plt.bar(x,y, color= "blue");
plt.xlabel("Gender");
plt.ylabel("Survived");
plt.title("Age VS Survived");
plt.show();
```
